Create a new Column with the following condition:\

>10 : Long Trip\
5-10: Medium Trip\
 <5 : Short Trip

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('My Uber Drives - 2016.csv')
data
```

```python
data['TRIP_CATEGORIES'] = np.where(data['MILES*'] < 5, 'SHORT TRIP', np.where(data['MILES*'] <= 10, 'MEDIUM TRIP', 'LONG TRIP'))
```
